### Sources of Market Failure and Relation to the Literature
-  **Pecuniary Externalities**
> - In this paper, the sources of market failure are pecuniary externalities.
 - *$Competitive\ bankers$*
    - Complete competitive market
    - Competitive bankers are price taker,who choose their lending plans so as to maximize profits,under (price dependent) leverage constraint imposed by investors' Participation constraint
    - Bank and representative investors sign loan contract,but bankers cannot pledge their entire future payoffs to their depositors/investors,result from Financial friction .
    - 
 - *$social\  planner$*
      -  capital prices are not given, but respond to changes in the lending plans of banks, which are all exposed to the same aggregate shock and thus are perfectly correlated. 
      - **capital price is such that individual bankers are indifferent between shifting one additional unit of  borrowing capacities across states at the equilibrium prices**
      -  Banks lend too much in the good state because they do not internalize the impact of their lending decisions on capital prices and thus on banks’capital constraints.
      - By forcing all banks to lend less in the best state, the planner would decrease the price of capital in this state and would generate strictly higher returns for investors in this state than their alternative investment opportunities in the bond market. In turn, investors would be willing to lend more in the other state, such that in expectation.
      - Net effect excessively favors lending in good times at the expense of lending in bad times.

- **Financial Contracting and Aggregate Economic Activities**
> - First, banks face binding leverage constraints in all states of the world.
 - Second, agents can enter complete contracts with lenders and can ex ante allocate their borrowing capacity flexibly across states of the world. Hence, the effect of pecuniary externalities is a priori unclear.
 - under plausible conditions, competitive banks allocate too much borrowing capacity to good times ,while fewer in bad times. Then, imposing a (stricter) capital requirement in good states corrects the misallocation of the borrowing capacity and increases expected output and social welfare

- **Counter-Cyclical Capital Requirements**
> -  Repullo and Suarez (2013) and Repullo (2013) show that without cyclical adjustments of (risk-sensitive) capital requirements, aggregate investments may decline excessively (credit crunches:信贷紧缩) after negative shocks to bank capital. 
 - Our mechanism is symmetric: counter-cyclical capital requirements correct both credit booms in good times and credit crunches in bad times.

- **Capital and Liquidity Regulation**
> - Capital requirements reduce risk-taking, but they also reduce liquidity.
  - Kara and Ozsoy (2016) study a model with fire-sales externalities in which banks choose the amount of $liquid \ assets$ and the amount of $risky \ assets$. They show that only the use of regulatory tools for both capital  and liquidity can achieve the welfare optimum.
  - Current paper only focuses on  regulatory tool, namely counter-cyclical capital requirements

- **Modeling and Empirics**
 >   - Assume that bankers cannot pledge their entire future pay-offs to their depositors/investors, as they may divert some fraction of their assets.
    - The present model provides a theoretical explanation why such regulation can reduce volatility of bank lending. 
    - Second, it replicates well-known stylized facts on credit fluctuations and bank balance sheets, as established comprehensively in Adrian et al. (2012). 
    - For instance, bank leverage is pro-cyclical and bond financing to non-financial firms substitutes for declining bank lending in downturns.

### The timing of the model is as follows:
- The total physical capital stock of the economy is normalized to 1. Bankers own (on aggregate) a fraction E of this capital. E can be interpreted as the aggregate capitalization of the banking sector. Investors own the remaining capital 1 − E. Entrepreneurs own the firms in the frictionless sector and bank-dependent sector. All agents are risk-neutral and endeavor to maximize their expected consumption at date 2.

- **at date 0, all agents can trade on contingent markets for capital. There is a market for each state, i.e. financial markets are complete. Thanks to these markets, banks can obtain a state dependent capital endowment, denoted by $e_s$ in states $s = h, l$**
- **at date 1, all agents observe the aggregate state s. A typical banker borrows $k_s − e_s > 0$ units of capital from investors/depositors, and invests ks units of capital. The aggregate size of the banking sector (the integral of ks over all banks) is denoted by $K_s$. We interpret $K_s$. as the aggregate volume of credit to the economy in state s. The remaining amount of capital $1-K_s=X_s$. is invested in the traditional sector.**
- **at date 2, output in the banking sector, RsKs, is shared between investors and bankers. More specifically, the investors who, at date 1, have lent  $k_s − e_s $ units of capital to a typical bank receive  $p_s(k_s − e_s )$ units of consumption at date 2. Since consumption is taken as a numeraire, the rate of return on banks’ deposits, namely  $p_s$, can also be interpreted as the price of capital in state s, as it corresponds to the number of units of consumption that are delivered at date 2 in exchange for one unit of capital at date 1. The output in the traditional sector $F (1 − K_s) $accrues to investors. The total output in the economy is $F (1 − K_s) + R_sK_s$**
